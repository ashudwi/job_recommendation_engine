```python
import pandas as pd
import numpy as np

#importing the data
tr_data= pd.read_csv("TrainData.csv", sep=',', header='infer')
tr_data.head()
```

```python
#checking the number of user and their ratings
print(tr_data.shape)
tr_data.groupby(['Rating'])['JobID'].count()
```

```python
#removing records who gave ratings to less than 2 job id
import collections
userRatingsCount = collections.defaultdict(int)
for i in range(len(tr_data)):
    userRatingsCount[tr_data["ApplicantID"][i]] += 1
    ## Removing users who have given only one rating
for key,val in userRatingsCount.items():
    if(val == 1):
        tr_data = tr_data.drop(tr_data.index[np.where(tr_data['ApplicantID'] == key)])
```

```python
#calculating the number of ratings each user gave

pd_matrix = tr_data.pivot(index='ApplicantID', columns='JobID',values='Rating')
pd_matrix = pd_matrix.fillna(0)

number_of_ratings_each_user_gave = (pd_matrix != 0).astype(int).sum(axis=1).reset_index()
number_of_ratings_each_user_gave.columns = ['ApplicantId','non_zero']
number_of_ratings_each_user_gave =number_of_ratings_each_user_gave.sort_values(by='non_zero', ascending=False)
number_of_ratings_each_user_gave.head()
```

```python
#setting dataframe for svd
pd_matrix = pd_matrix.replace(0,np.nan).reset_index()
pd_matrix.head()
```

```python
#creating dataframe for recommendations
jobid = pd_matrix.drop('ApplicantID',1)
jobid=jobid.T.fillna(jobid.mean(axis=1)).T
jobid.head()
```

```python
#creating svd matrix
U,S,V = np.linalg.svd(jobid)
U.shape
```

```python
user = 6945
user_svd = U[np.where(pd_matrix['ApplicantID'] == user)]
user_svd.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#creating cosine similarity of the user with all other users in U matrix and selecting top 30
similarusers = map(lambda x: cosine_similarity(user_svd, x.reshape(1,-1)), U)
top = sorted(range(len(similarusers)), key=lambda i: similarusers[i], reverse=True)[:30]
top
```

```python
## Retrieving job ids applied by the top 30 most similar users to the selected user
recommended = pd.DataFrame(columns = ['ApplicantID','JobID','data'])
count = 0
for i in top:
    sameusers = pd_matrix.loc[[i]].dropna(axis = 1).drop('ApplicantID', axis = 1).columns
#     print(similarUserjobIds)
    for j in sameusers:
        ## Storing only that jobId which was not already data in the dataframe
        if((recommended.loc[recommended['JobID'] == j]).empty == True):
            #jobRecommendations_SVD.set_value(count, 'ApplicantID',user)
            recommended.set_value(count,'JobID', j)
            count += 1
```

```python
#printing the recommended jobs
recommended['ApplicantID'] = user
```

```python
job_data = pd.read_csv("job_data.csv")
recommended = recommended.merge(job_data, left_on='JobID',right_on='Job.ID', how='left')
```

```python
recommended = recommended[pd.notnull(recommended['text'])]
```

```python
recommended = recommended[['ApplicantID', 'Job.ID','text']]
recommended
```
